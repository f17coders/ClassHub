```python
import pandas as pd
import json
```

```python
with open("../work_classification/cat2svc.json", "r", encoding="utf-8") as f:
    cat_json = json.load(f)
cat_json
```

```python
lecture_inflearn = pd.read_csv("./lecture_inflearn_new.csv")
lecture_inflearn["category_name"] = lecture_inflearn["category_name"].apply(lambda x:cat_json[x])
lecture_inflearn = lecture_inflearn.drop(columns=["gpt_review_good","gpt_review_bad"])
lecture_inflearn["summary"] = lecture_inflearn["summary"].fillna("")
lecture_inflearn["description_summary"] = lecture_inflearn["description_summary"].fillna("")

lecture_inflearn.to_csv("./lecture_inflearn.csv",index=False)
```

# tag

```python
with open("../work_classification/preprocessed2servicetag.json", "r", encoding="utf-8") as f:
    tag_json = json.load(f)
tag_json
```

```python
tag_inflearn = pd.read_csv("./tag_inflearn.csv")
tag_inflearn = tag_inflearn[tag_inflearn["tags"].isna()==False]

site_lecture_id_list = []
tags_list = []

print(tag_inflearn.isna().sum())

for _, row in tag_inflearn.iterrows():
    tags = row["tags"].split("||")
    id = row["site_lecture_id"]
    tags_list.extend(tags)
    site_lecture_id_list.extend([id for i in range(len(tags))])

tag_inflearn = pd.DataFrame({
    "site_lecture_id" : site_lecture_id_list,
    "tag_name" : tags_list
})

temp = list(tag_inflearn["tag_name"].values)
temp = list(map(lambda x:"".join(x.strip().split()).lower(), temp))

for tag in temp:
    if tag not in tag_json:
        print("not in")

temp = list(map(lambda x:tag_json[x],temp))
print(temp)

tag_inflearn["tag_name"] = temp

tag_inflearn.to_csv("./tag_inflearn.csv",index=False)
```
