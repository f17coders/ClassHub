```python
import pandas as pd
import json
```

```python
with open("../work_classification/cat2svc.json", "r", encoding="utf-8") as f:
    cat_json = json.load(f)
cat_json
```

```python
lecture_goorm = pd.read_csv("./lecture_goorm_new.csv")
lecture_goorm = lecture_goorm.drop(columns=["gpt_review_good","gpt_review_bad"])
lecture_goorm["summary"] = lecture_goorm["summary"].fillna("")

lecture_goorm["price_original"] = lecture_goorm["price_original"].fillna(0)
lecture_goorm["price_sale"] = lecture_goorm["price_sale"].fillna(0)
lecture_goorm["description_summary"] = lecture_goorm["description_summary"].fillna("")

lecture_goorm['price_original'] = lecture_goorm['price_original'].astype(int)
lecture_goorm['price_sale'] = lecture_goorm['price_sale'].astype(int)

temp = list(lecture_goorm["category_name"].values)
temp = list(map(lambda x:' '.join(x.split("-")), temp))
for cat in temp:
    if cat not in cat_json:
        print(f"not in {cat}")
lecture_goorm["category_name"] = temp
lecture_goorm["category_name"] = lecture_goorm["category_name"].apply(lambda x:cat_json[x])


print(lecture_goorm.isna().sum())

lecture_goorm.to_csv("./lecture_goorm.csv",index=False)


# print(lecture_goorm[lecture_goorm["price_original"].isna()]) # 무료강의
# print(lecture_goorm[lecture_goorm["price_sale"].isna()]) # 무료강의
# print(lecture_goorm.loc[148])
# lecture_goorm
```

# tag

```python
with open("../work_classification/preprocessed2servicetag.json", "r", encoding="utf-8") as f:
    tag_json = json.load(f)
tag_json
```

```python
tag_goorm = pd.read_csv("./tag_goorm.csv")
tag_goorm = tag_goorm[tag_goorm["tags"].isna()==False]

print(tag_goorm[tag_goorm["tags"]==""])
print(tag_goorm.isna().sum())

site_lecture_id_list = []
tags_list = []

for _, row in tag_goorm.iterrows():
    tags = row["tags"].split("||")
    id = row["site_lecture_id"]
    tags_list.extend(tags)
    site_lecture_id_list.extend([id for i in range(len(tags))])

tag_goorm = pd.DataFrame({
    "site_lecture_id" : site_lecture_id_list,
    "tag_name" : tags_list
})

temp = list(tag_goorm["tag_name"].values)
temp = list(map(lambda x:"".join(x.strip().split()).lower(), temp))

for tag in temp:
    if tag not in tag_json:
        print("not in")

temp = list(map(lambda x:tag_json[x],temp))

tag_goorm["tag_name"] = temp

tag_goorm.to_csv("./tag_goorm.csv",index=False)
```
