```python
import json
import pandas as pd
```

```python
with open('../udemy/ko_list.json', 'r', encoding="utf-8") as file:
    ko_list = list(json.load(file))
ko_list
```

```python
udemy_lectures = pd.read_csv("../udemy/lectures/new_udemy_lectures_dummy.csv")
udemy_lectures
```

```python
ko_list = pd.DataFrame({"site_lecture_id":list(map(lambda x:f"u{x}",ko_list))})
```

```python
pd.merge(udemy_lectures, ko_list, how="inner", left_on="site_lecture_id", right_on="site_lecture_id").to_csv("../udemy/lectures/ko_udemy_lectures.csv",index=False)
```
