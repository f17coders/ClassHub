# gpt 전처리

```python
gpt_reviews = pd.read_csv("../udemy/gpt/full_gpt.csv")
gpt_reviews
```

```python
len("aaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaaa")
```

```python
from tqdm import tqdm
new_one = []
for org in tqdm(list(map(str,gpt_reviews["gptReviews"].to_list()))):
    lines = org.split("\n")
    for i in range(len(lines)):
        line = lines[i]
        tokens = line.split(" ")
        for j in range(len(tokens)):
            token = tokens[j]
            if(len(token)>80):
                new_token = ""
                for idx in range(0,len(token),80):
                    new_token += token[idx:idx+80]+" "
                new_token = new_token.rstrip()
                tokens[j] = new_token
        lines[i] = ' '.join(tokens)
    new_one.append('\n'.join(lines))
    # print('\n'.join(lines))
    # break
```

```python
gpt_reviews["processed_gpt"] = new_one
gpt_reviews
```

```python
gpt_reviews["length"] = gpt_reviews["processed_gpt"].apply(lambda x:len(x))
gpt_reviews.sort_values("length",ascending=False)
```

```python
gpt_reviews["site_lecture_id"] = gpt_reviews["site_lecture_id"].apply(lambda x:f"u{x}")
gpt_reviews
```

```python
gpt_reviews.to_csv("../udemy/gpt/processed_full_gpt.csv",index=False)
```

# 한국어 필터링과 table 완성

```python
udemy_lectures = pd.read_csv("../udemy/lectures/ko_udemy_lectures.csv")
udemy_lectures
```

```python
with open('../udemy/ko_list.json', 'r', encoding="utf-8") as file:
    ko_list = list(json.load(file))
ko_list
ko_df = pd.DataFrame({"site_lecture_id":list(map(lambda x:f"u{x}",ko_list))})
ko_df
```

```python
ko_udemy_lectures = pd.merge(udemy_lectures, ko_df, how="inner", left_on="site_lecture_id", right_on="site_lecture_id")
ko_udemy_lectures
```

```python
gpt_reviews[["site_lecture_id","processed_gpt"]]
```

```python
merged_temp = pd.merge(ko_udemy_lectures, gpt_reviews[["site_lecture_id","processed_gpt"]], how="left", left_on="site_lecture_id", right_on="site_lecture_id")
merged_temp
```

```python
merged_temp["processed_gpt"] = merged_temp["processed_gpt"].fillna("리뷰의 개수가 적어 GPT리뷰를 생성 불가합니다.")
merged_temp
```

```python
merged_temp = merged_temp.drop(columns=["gpt_review_good","gpt_review_bad"])
merged_temp = merged_temp.rename(columns = {"processed_gpt":"gpt_review"})
merged_temp["description_detail"] = merged_temp["site_lecture_id"].apply(lambda x:f"{x}.html")
merged_temp
```

```python
merged_temp.to_csv("../udemy/lectures/new_udemy_lectures_dummy.csv", index=False)
```
