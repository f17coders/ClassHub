# goorm, inflearn gpt리뷰 생성 위한 테이블 만들기

```python
review_inflearn = pd.read_csv("../inflearn/review_inflearn.csv")
review_goorm = pd.read_csv("../goorm/review_goorm.csv")

reviews = pd.concat([review_inflearn,review_goorm]).reset_index(drop=True)
reviews["content"] = reviews["content"].apply(lambda x: "" if len("".join(str(x).strip().split()))<7 else str(x))
reviews = reviews.reset_index(drop=True)
reviews = reviews[reviews['content'].str.len()>7]
reviews["content"] = reviews['content'].apply(lambda x: f"{x}." if x[-1]!="." else x)


# content 열의 문자열 길이 계산하여 새로운 열 추가
reviews['content_length'] = reviews['content'].apply(len)

# print(reviews)

# lecture_id로 그룹화하여 content_length의 누적합 계산
reviews["cumsum"] = reviews.groupby('id')['content_length'].cumsum()
reviews = reviews[reviews["cumsum"]<3500]

reviews = reviews.groupby('id')["content"].apply(lambda x: ' '.join(x)).reset_index()
reviews["prompt"] = reviews["content"].apply(lambda x:f"지시문 : 다음은 하나의 it 인터넷 강의의 리뷰들입니다. 장점과 단점에 해당하는 요약을 각각 30자 이내로 생성해줘. Please write in korean\n{x}")
reviews = reviews.rename(columns={"id":"site_lecture_id"})

reviews.to_csv("./prompt_integrated.csv",index=False)
reviews
```
